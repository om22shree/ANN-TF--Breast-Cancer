# Imports


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Tensorflow import and initiation
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

# DataFrame Creation

```python
breast_data = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_data.data, columns = breast_data.feature_names)
```

```python
data_frame.head()
```

```python
## Adding targets to our dataframe.
data_frame['label'] = breast_data.target
```

```python
data_frame.tail()
```

```python
## Statistical summary and null checks
print(data_frame.shape)
data_frame.info()
data_frame.isnull().sum()
data_frame.describe()
```

```python
## Check the taget variable distribution.
data_frame["label"].value_counts()
## Data is imbalanced in the favour of 1, imbalance is within tolerance.
```

### 1 --> Benign
### 0 --> Malignant

# Train Test Split

```python
x = data_frame.drop(columns="label", axis=1)
y = data_frame["label"]
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                    test_size=0.2, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

# Data Standardization

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

# Building the neural network

```python
# NN Layer set-up
model = keras.Sequential([
    # Input layer : 30 features
    keras.layers.Flatten(input_shape=(30, )),
    # Hidden layer : 20 hidden layers
    keras.layers.Dense(20, activation="relu"),
    # Output layer : 2 possible classes(benign/malignant)
    keras.layers.Dense(2, activation="sigmoid")
])
```

```python
# Compiling the NN model
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=['accuracy'])
```

```python
# Training the NN
history = model.fit(x_train, y_train, validation_split=0.1, epochs=10)
```

# Model analysis

```python
# Visualizing the loss and error
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('NN Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
# Visualizing the loss and error
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('NN Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['training data', 'validation data'], loc='upper right')
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(accuracy)
```
